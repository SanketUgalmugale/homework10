```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('imdb_dataset.csv')
```

```python
df
```

```python
# Handling invalid values: convert "xx" in the height(cm) column to NaN
df['height(cm)'] = df['height(cm)'].replace('xx', np.nan)
```

```python
# Split 'age_sex' column into 'age' and 'sex' columns
df[['age', 'sex']] = df['age_sex'].str.extract('(\\d+)([MF])')
```

```python
# Drop the original 'age_sex' column
df.drop('age_sex', axis=1, inplace=True)
```
